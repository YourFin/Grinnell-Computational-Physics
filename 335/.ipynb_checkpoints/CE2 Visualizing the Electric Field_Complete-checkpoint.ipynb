```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E(q, r0, x, y):
    """Return the electric field vector E=(Ex,Ey) due to charge q at r0."""
    den = np.hypot(x-r0[0], y-r0[1])**3
    return q * (x - r0[0]) / den, q * (y - r0[1]) / den

#Define constants, locations, and charges
e_0 = 8.854*(10**-12) #epsilon_0 in F/m

q1_loc = (0, 0)
q1 = 1

q2_loc = (0, 1)
q2 = -1

#Create set of points
res = 32
x = np.linspace(-1, 2, res)
y =np.linspace(-1, 2, res)
X, Y = np.meshgrid(x, y)

Ex, Ey = np.meshgrid(x, y)

#Find E at each 
for i in range(res):
    for j in range(res):
        if (i, j) == q1_loc or (i, j) == q2_loc:
            """Skip over the points themselves to avoid errors"""
            Ex[i, j] = 0
            Ey[i, j] = 0
        else:
            e1x, e1y = E(q1, q1_loc, x[i], y[j])
            e2x, e2y = E(q2, q2_loc, x[i], y[j])
            Ex[i,j] = e1x+e2x
            Ey[i,j] = e1y+e2y

#Normalize vectors, multiplying them to make them easier to see
Ex = 10*Ex/np.sqrt(Ex**2 + Ey**2)
Ey = 10*Ey/np.sqrt(Ex**2 + Ey**2)


#Plot Vector Grid
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.quiver(X, Y, Ex, Ey)

ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_xlim(-1,2)
ax1.set_ylim(-1,2)
ax1.set_aspect('equal')
plt.show()


#Plot Contours
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.streamplot(x, y, Ex, Ey, linewidth=1, density=2, arrowstyle='->', arrowsize=1.5)

ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_xlim(-1,2)
ax2.set_ylim(-1,2)
ax2.set_aspect('equal')
plt.show()
```
